```python
from deductor_parser import DeductorParser
from neuronet import ActivePerceptron
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import scipy.ndimage as ndimage
```

```python
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (15,10)
```

```python
parser = DeductorParser('гео 2.ded')
```

```python
parser.setdocument('Текстовый файл (C:\\Users\\Aleksandr\\Desktop\\сканы\\data.csv)')
```

```python
parser.setneuronet('Нейросеть [2 x 3 x 3 x 3 x 3]', parse=True)
```

```python
data = parser.neurodata
```

```python
nw = ActivePerceptron()
nw.loader(data)
```

```python
n_min, n_max = data['inp_norm'][0]
e_min, e_max = data['inp_norm'][1]
```

```python
n_steps = 200
e_steps = 200
```

```python
def drawPlot(n):
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    X = np.arange(n_min, n_max, (n_max - n_min) / n_steps)
    Y = np.arange(e_min, e_max, (e_max - e_min) / e_steps) 
    X, Y = np.meshgrid(X, Y)
    Z = np.zeros([n_steps, e_steps])
    for i in range(len(Z)):
        for j in range(len(Z[0])):
            Z[i][j] = nw([X[i][j], Y[i][j]])[n - 1]
    Z = ndimage.gaussian_filter(Z, sigma=1.0, order=0)

    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False, rstride=1, cstride=1, alpha=1)

    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
    ax.view_init(azim=150)

    fig.colorbar(surf, shrink=0.5, aspect=5)
    ax.set_xlabel('N_coord')
    ax.set_ylabel('E_coord')
    ax.set_zlabel('Y' + str(n))
    plt.title('Y' + str(n))

    plt.show()
```

```python
drawPlot(1)
```

```python
drawPlot(2)
```

```python
drawPlot(3)
```
